```python
import numpy as np
import itertools
import matplotlib.pyplot as plt

N = 2
latt1 = np.arange(1, 9).reshape([N, N,2])
latt = np.ones(N**2*2).reshape([N, N,2])

rng = np.random.default_rng()

def randomstart():

    sites = itertools.product(range(N), range(N), range(2))

    for site in sites:

        latt[site] = 2*rng.integers(2)-1
        

    return None


def coldstart():
    latt[:] = 1
    return None

def moveup(xvec, d):
    #  xvec is mutable
    xvec[d] += 1
    xvec[d] = xvec[d] % N   # stays between [0, 1, ..., N-1]
    return None


def movedown(xvec, d):
    #  xvec is mutable
    xvec[d] -= 1
    xvec[d] = xvec[d] % N
    return None

def update(beta):

    sites = itertools.product(range(N), range(N), range(2))

    action = 0.

    for site in sites:

        *x, d = site

        # following M. Creutz
        # staples around link(1->4)
        #    dperp        6--5
        #    ^            |  |
        #    |            1--4
        #    |            |  |
        #    -----> d     2--3

        staplesum = 0.
        for dperp in range(2):
            if dperp != d:

                 # plaquette 1234
                #staple = latt[x[0], x[1], d]
                #C1 = latt[x[0], x[1], d]
                movedown(x, dperp)
                staple = latt[x[0], x[1], dperp]
                A = latt1[x[0], x[1], dperp]
                staple *= latt[x[0], x[1], d]
                B = latt1[x[0], x[1], d]
                moveup(x, d)
                staple *= latt[x[0], x[1], dperp]
                C = latt1[x[0], x[1], dperp]
                #print("plaquette1 ",  C,B,A, C1)
                print("plaquette ",  C,B,A)
                print("-----------")
                staplesum += staple

                # plaquette 4561
                moveup(x, dperp)
                staple = latt[x[0], x[1], dperp]
                D = latt1[x[0], x[1], dperp]
                moveup(x, dperp)
                movedown(x, d)
                staple *= latt[x[0], x[1], d]
                E = latt1[x[0], x[1], d]
                movedown(x, dperp)
                staple *= latt[x[0], x[1],  dperp]
                F = latt1[x[0], x[1],  dperp]
                #staple *= latt[x[0], x[1],  d]
                #F1 = latt[x[0], x[1],  d]
                #print("plaquette2 ", F1, F,E,D)
                print("plaquette ", F,E,D)
                print("-----------")
                print("-------------------------------------------------------------------------------------")
                print("-------------------------------------------------------------------------------------")
                staplesum += staple

        # calculate the Boltzmann weight
        print("staplesum ", staplesum)
        bplus = np.exp(beta*staplesum)
        bminus = np.exp(-beta*staplesum)
        bplus = bplus/(bplus+bminus)
        print("bplus ", bplus)

        # the heatbath algorithm
        r = rng.uniform()

        if r < bplus:
            print("r", r)
            latt[site] = 1
            action += staplesum
            print("action", action)
        else:
            latt[site] = -1
            action -= staplesum
            print("action", action)
            print("-------------------------------------------------------------------------------------")
            print("-------------------------------------------------------------------------------------")
        print("action", action)
    return 1. - action/N**2/2./6.

#latt
update(1)
```
