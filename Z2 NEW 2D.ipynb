```python
##works well!

##################################################
#        2-Dimensional Z2 Lattice Gauge          #
#   By Hariprashad Ravikumar [hari1729@nmsu.edu] #
#   For Dr. Engelhardt's meeting (Dec 22, 2022)  #
##################################################

import numpy as np
import numpy.random as rnd
from scipy import random
import itertools
import matplotlib.pyplot as plt


N = 100 # Size of the lattice
latt1 = np.ones((N, N), dtype=int)
latt2 = np.ones((N, N), dtype=int)

# verifying the plaquette
#start = 1
#end = 9 + 1
#latt11 = np.arange(start, end-1, 2).reshape([N, N])
#latt22 = np.arange(start+1, end, 2).reshape([N, N])



rng = np.random.default_rng()
def randomstart():
    sites = itertools.product(range(N), range(N))
    for site in sites:
        spin = rng.integers(2)
        if spin == 0:
            spin = -1
        latt1[site] = spin
        latt2[site] = spin
    return None


def coldstart():
    latt1[:] = 1
    latt2[:] = 1
    return None      

def action_axa_update(configuration1, configuration2, beta):
    action_total = 0.0
    count = 0.0
    # middle and right
    for i in range(N):
        for j in range(N):
            #plaquette middle
            left1 = configuration1[i,j]
            down1 = configuration2[i,j]
            right1 = configuration1[i,(j+1)%N] ##
            top1 = configuration2[(i-1)%N,j]
            
            #plaquette right
            r_left = right1 ##
            r_down = configuration2[i,(j+1)%N]
            r_right = configuration1[i,(j+2)%N]
            r_top = configuration2[(i-1)%N,(j+1)%N]
            
            staplesum_middle1 = ((top1*left1*down1))
            staplesum_right = ((r_down*r_right*r_top))
            #print(latt22[(i-1)%N,j],latt11[i,j],latt22[i,j])
            #print(latt22[i,(j+1)%N],latt11[i,(j+2)%N],latt22[(i-1)%N,(j+1)%N])
            staplesum_middle_right = (staplesum_middle1+staplesum_right) 
            count += 2
            bplus1 = np.exp(beta*staplesum_middle_right)
            bminus1 = np.exp(-beta*staplesum_middle_right)
            bplus1 = bplus1/(bplus1+bminus1)
            r1 = rng.uniform()
            if r1 < bplus1:
                configuration1[i,(j+1)%N] = 1
                action_total += staplesum_middle_right
            else:
                configuration1[i,(j+1)%N] = -1
                action_total -= staplesum_middle_right
    
    # middle and top
    for i in range(N):
        for j in range(N):
            #plaquette middle
            left3 = configuration1[i,j] 
            down3 = configuration2[i,j]
            right3 = configuration1[i,(j+1)%N]
            top3 = configuration2[(i-1)%N,j] ##
            
            #plaquette top
            t_left = configuration1[(i-1)%N,j]
            t_down = top3 ##
            t_right = configuration1[(i-1)%N,(j+1)%N]
            t_top = configuration2[(i-2)%N,j]
            
            staplesum_middle3 = ((left3*down3*right3))
            staplesum_top = ((t_right*t_top*t_left))
            #print(latt11[i,j] ,latt22[i,j], latt11[i,(j+1)%N])
            #print(latt11[(i-1)%N,(j+1)%N],latt22[(i-2)%N,j],latt11[(i-1)%N,j])
            staplesum_middle_top = (staplesum_middle3+staplesum_top) 
            count += 2
            bplus3 = np.exp(beta*staplesum_middle_top)
            bminus3 = np.exp(-beta*staplesum_middle_top)
            bplus3 = bplus3/(bplus3+bminus3)
            r3 = rng.uniform()
            if r3 < bplus3:
                configuration2[(i-1)%N,j] = 1
                action_total += staplesum_middle_top
            else:
                configuration2[(i-1)%N,j] = -1
                action_total -= staplesum_middle_top
    confi1 = configuration1
    confi2 = configuration2
    return -action_total/(N*N), confi1, confi2

def wilson_axa(configuration1, configuration2, beta):
    Wilson_loop__axa = 0.0
    count = 0.0
    for i in range(N):
        for j in range(N):
            left = configuration1[i,j]
            down = configuration2[i,j]
            right = configuration1[i,(j+1)%N]
            top = configuration2[(i-1)%N,j]
            Wilson_loop__axa += ((left*down*right*top))
            count += 1
    return  Wilson_loop__axa/(count)


def wilson_ax2a(configuration1, configuration2, beta):
    Wilson_loop__ax2a = 0.0
    count = 0.0
    for i in range(N):
        for j in range(N):
            left1 = configuration1[i,j]
            left2 = configuration1[(i+1)%N,j]
            down = configuration2[(i+1)%N,j]
            right2 = configuration1[(i+1)%N,(j+1)%N]
            right1 = configuration1[i,(j+1)%N]
            top = configuration2[(i-1)%N,j]
            #print(left1,left2,down,right2,right1,top)
            Wilson_loop__ax2a += ((left1*left2*down*right2*right1*top))
            count += 1
    return  Wilson_loop__ax2a/(count)


def wilson_2axa(configuration1, configuration2, beta):
    Wilson_loop__2axa = 0.0
    count = 0.0
    for i in range(N):
        for j in range(N):
            left = configuration1[i,j]
            down1 = configuration2[i,j]
            down2 = configuration2[i,(j+1)%N]
            right = configuration1[i,(j+2)%N]
            top2 = configuration2[(i-1)%N,(j+1)%N]
            top1 = configuration2[(i-1)%N,j]
            #print(left,down1,down2,right,top2,top1)
            Wilson_loop__2axa += ((left*down1*down2*right*top2*top1))
            count += 1
    return  Wilson_loop__2axa/(count)


def wilson_2ax2a(configuration1, configuration2, beta):
    Wilson_loop__2ax2a = 0.0
    count = 0.0
    for i in range(N):
        for j in range(N):
            left1 = configuration1[i,j]
            left2 = configuration1[(i+1)%N,j]
            down1 = configuration2[(i+1)%N,j]
            down2 = configuration2[(i+1)%N,(j+1)%N]
            right2 = configuration1[(i+1)%N,(j+2)%N]
            right1 = configuration1[i,(j+2)%N]
            top2 = configuration2[(i-1)%N,(j+1)%N]
            top1 = configuration2[(i-1)%N,j]
            #print(left1,left2,down1,down2,right2,right1,top2,top1)
            Wilson_loop__2ax2a += ((left1*left2*down1*down2*right2*right1*top2*top1))
            count += 1
    return  Wilson_loop__2ax2a/(count)


def wilson_ax3a(configuration1, configuration2, beta):
    Wilson_loop__ax3a = 0.0
    count = 0.0
    for i in range(N):
        for j in range(N):
            left1 = configuration1[i,j]
            left2 = configuration1[(i+1)%N,j]
            left3 = configuration1[(i+2)%N,j]
            down = configuration2[(i+2)%N,j]
            right3 = configuration1[(i+2)%N,(j+1)%N]
            right2 = configuration1[(i+1)%N,(j+1)%N]
            right1 = configuration1[i,(j+1)%N]
            top = configuration2[(i-1)%N,j]
            #print(left1,left2,left3,down,right3,right2,right1,top)
            Wilson_loop__ax3a += ((left1*left2*left3*down*right3*right2*right1*top))
            count += 1
    return  Wilson_loop__ax3a/(count)

def wilson_3ax3a(configuration1, configuration2, beta):
    Wilson_loop__3ax3a = 0.0
    count = 0.0
    for i in range(N):
        for j in range(N):
            left1 = configuration1[i,j]
            left2 = configuration1[(i+1)%N,j]
            left3 = configuration1[(i+2)%N,j]
            down1 = configuration2[(i+2)%N,j]
            down2 = configuration2[(i+2)%N,(j+1)%N]
            down3 = configuration2[(i+2)%N,(j+2)%N]
            right3 = configuration1[(i+2)%N,(j+3)%N]
            right2 = configuration1[(i+1)%N,(j+3)%N]
            right1 = configuration1[i,(j+3)%N]
            top3 = configuration2[(i-1)%N,(j+2)%N]
            top2 = configuration2[(i-1)%N,(j+1)%N]
            top1 = configuration2[(i-1)%N,j]
            #print(left1,left2,left3,down1,down2,down3,right3,right2,right1,top3,top2,top1)
            Wilson_loop__3ax3a += ((left1*left2*left3*down1*down2*down3*right3*right2*right1*top3*top2*top1))
            count += 1
    return  Wilson_loop__3ax3a/(count)




def expectation_axa(beta_test,N_0,MC_steps):
    beta_arr = np.ones(MC_steps)*beta_test
    exp_action_axa = np.zeros(len(beta_arr))
    exp_axa = np.zeros(len(beta_arr))
    exp_ax2a = np.zeros(len(beta_arr))
    exp_2ax2a = np.zeros(len(beta_arr))
    exp_ax3a = np.zeros(len(beta_arr))
    exp_3ax3a = np.zeros(len(beta_arr))
    count = 0.0
    for i, beta  in enumerate(beta_arr):
        action, config1, config2 = action_axa_update(latt1, latt2, beta)
        Wilson_loop_axa = wilson_axa(config1, config2, beta)
        Wilson_loop_ax2a = wilson_ax2a(config1, config2, beta)
        Wilson_loop_2ax2a = wilson_2ax2a(config1, config2, beta)
        Wilson_loop_ax3a = wilson_ax3a(config1, config2, beta)
        Wilson_loop_3ax3a = wilson_3ax3a(config1, config2, beta)
        count += 1 
        if count > N_0:
            exp_action_axa[i] += action/count
            exp_axa[i] += Wilson_loop_axa/count
            exp_ax2a[i] += Wilson_loop_ax2a/count
            exp_2ax2a[i] += Wilson_loop_2ax2a/count
            exp_ax3a[i] += Wilson_loop_ax3a/count
            exp_3ax3a[i] += Wilson_loop_3ax3a/count
    return abs(exp_action_axa), abs(exp_axa), abs(exp_ax2a), abs(exp_2ax2a), abs(exp_ax3a), abs(exp_3ax3a)




beta_value = 1
n_0 = 2
mc_steps = 100


exp_action_axa, exp_axa, exp_ax2a, exp_2ax2a, exp_ax3a, exp_3ax3a = expectation_axa(beta_value,n_0,mc_steps)


plt.subplots(figsize=(16, 9))
#plt.plot(exp_action_axa, color='blue',label= 'Action axa' )
plt.plot(exp_axa, color='blue',label= 'Wilson loop axa' )
plt.plot(exp_ax2a, color='green',label= 'Wilson loop ax2a' )
plt.plot(exp_2ax2a, color='red',label= 'Wilson loop 2ax2a' )
plt.plot(exp_ax3a, color='magenta',label= 'Wilson loop ax3a' )
plt.plot(exp_3ax3a, color='black',label= 'Wilson loop 3ax3a' )
plt.xlabel("MC_UPDATES", fontsize=18)
plt.ylabel("EXPECTATION VALUE OF WILSON LOOPS", fontsize=18)
plt.title("EXPECTATION VALUE  Vs. MC_UPDATES    (Size $%.f^2$)"%(N), fontsize=18)
plt.xlim([n_0, mc_steps])
plt.legend( fontsize=18)
plt.show()


```
